```python
import pandas as pd
from datetime import datetime, date
import requests
import json
```

```python
with open('.chave.txt', 'r') as file:
    chave_string = file.read().strip()
```

```python
## GET FLIGHTS
url = "https://google-flights2.p.rapidapi.com/api/v1/searchFlights"
headers = {
    "x-rapidapi-key": chave_string,
    "x-rapidapi-host": "google-flights2.p.rapidapi.com"
}
params = {
    "departure_id": "JFK",
    "arrival_id": "POA",
    "outbound_date": "2025-12-26",
    "return_date": "2026-01-06",
    "travel_class": "ECONOMY",
    "adults": "1",
    "show_hidden": "1",
    "currency": "USD",
    "language_code": "en-US",
    "country_code": "US",
    "search_type": "best"
}

response = requests.get(url, headers=headers, params=params)
print(response.status_code)
print(response.text)
json_output = response.text

with open(f'json_output_{date.today()}.txt', 'w') as f: f.write(json_output)
```

```python

# Replace this with your actual JSON string
json_str = json_output

# Parse the JSON string
result = json.loads(json_str)

# Explode into variables
status = result['status']
message = result['message']
timestamp = result['timestamp']
data = result['data']

itineraries = data['itineraries']
topFlights = itineraries['topFlights']
otherFlights = itineraries['otherFlights']
priceHistory = data['priceHistory']
price_summary = priceHistory['summary']
price_history = priceHistory['history']
```

```python
## Top Flights DataFrame

df_top_flights = pd.DataFrame(topFlights)
df_top_flights
```
